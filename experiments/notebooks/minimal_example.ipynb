# Minimal DataFrameSampler example

Load a small tabular dataset, inspect it, fit `DataFrameSampler`, and inspect generated rows.

```python
from sklearn.datasets import fetch_openml

from dataframe_sampler import DataFrameSampler
```

```python
adult = fetch_openml("adult", version=2, as_frame=True, parser="auto")
dataframe = adult.frame[
    [
        "age",
        "workclass",
        "education",
        "marital-status",
        "occupation",
        "hours-per-week",
        "class",
    ]
].dropna().sample(n=100, random_state=42).reset_index(drop=True)

dataframe.head()
```

```python
sampler = DataFrameSampler().fit(dataframe)
generated = sampler.generate(n_samples=10)

generated.head()
```
