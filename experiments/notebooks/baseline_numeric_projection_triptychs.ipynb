# Baseline Numeric Projection Triptychs

Choose a configured dataset and compare the numeric projection triptych for DataFrameSampler and the low-setup baseline methods. The same fitted DataFrameSampler latent view is used to project every generated dataframe, so the panels are comparable across methods.


```python
from pathlib import Path
import sys

try:
    get_ipython().run_line_magic("load_ext", "autoreload")
    get_ipython().run_line_magic("autoreload", "2")
except NameError:
    pass

ROOT = Path.cwd().resolve()
while not (ROOT / "pyproject.toml").exists() and ROOT != ROOT.parent:
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from IPython.display import display

from experiments.datasets import DATASET_CONFIGS
from experiments.numeric_projection import plot_numeric_projection_triptychs_for_methods
from experiments.workflow import load_dataset, working_dataframe
```

```python
DATASET_NAME = "titanic"
N_SAMPLES = 400
REDUCER = "umap"
INCLUDE_CTGAN = True
SAVE_FIGURES = False
```

```python
CONFIG = DATASET_CONFIGS[DATASET_NAME]
dataframe = load_dataset(CONFIG, root=ROOT)
work = working_dataframe(dataframe, CONFIG)
work.head()
```

```python
figures_dir = ROOT / "experiments" / "figures" if SAVE_FIGURES else None
figures, status = plot_numeric_projection_triptychs_for_methods(
    work,
    CONFIG,
    n_samples=min(N_SAMPLES, len(work)),
    include_ctgan=INCLUDE_CTGAN,
    reducer=REDUCER,
    figures_dir=figures_dir,
)
status
```

```python
for method, fig in figures.items():
    display(fig)
```
